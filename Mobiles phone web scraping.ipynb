```python
name=soup.find_all('a',class_="a-link-normal a-text-normal")
name[0:4]
```

```python
for i in name:
    Name_product.append(i.get_text().replace("\n",""))
```

```python
Name_product[0:4]
```

```python
print(len(Name_product))
```

```python
price=soup.find_all('span',class_="a-price-whole")
price[0:4]
```

```python
for i in price:
    Price_product.append(i.get_text().replace("\n",""))
```

```python
Price_product[0:5]
```

```python
print(len(Price_product))
```
