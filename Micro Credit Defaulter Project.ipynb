# Ckecking Skewness

```python
for col in micro.describe().columns:
    sns.distplot(micro[col],color='B')
    plt.show()
```

```python
micro.skew()
```

```python
#Treating skewness via square root and cube root
```

```python
for col in micro.skew().index:
    if col in micro.describe().columns:
        if micro[col].skew()>0.55:
            micro[col]=np.sqrt(micro[col])
        if micro[col].skew()<-0.55:
            micro[col]=np.cbrt(micro[col])
```

# Model Training

```python
micro.head()
```

```python
#Splitting the data into input and out variable
```

```python
x=micro.drop(columns=['label'],axis=1)
```

```python
x.head()
```

```python
y=micro.label
```

```python
y.head()
```

```python
x.shape
```

```python
y.shape
```

```python
#Scaling in input variable
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=44)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#importing the model libraries

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

#importing Boosting models
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier

#import Error Metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
model=[LogisticRegression(),GaussianNB(),DecisionTreeClassifier(),RandomForestClassifier(),
      AdaBoostClassifier(),GradientBoostingClassifier(),BaggingClassifier(),ExtraTreesClassifier()]


for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    
    print("Accuracy score of ",m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('**************************************')
    print('\n')
```

As we can see that which model has performed better

```python
#Cross validate the models
```

```python
model=[LogisticRegression(),GaussianNB(),DecisionTreeClassifier()]

for m in model:
    score=cross_val_score(m,x,y,cv=15,scoring='accuracy')
    print('model:',m)
    print('score:',score)
    print('mean score:',score.mean())
    print('standard deviation:',score.std())
    print('*********************************')
    print('\n')
```

```python
# find out best parameter in GaussianNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_train,y_train)
predgnb=gnb.predict(x_test)
print(accuracy_score(y_test,predgnb))
print(confusion_matrix(y_test,predgnb))
print(classification_report(y_test,predgnb))
```

```python
#find out best parameter in LogisticRegression
```

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
lg.score(x_train,y_train)
predlg=lg.predict(x_test)
print(accuracy_score(y_test,predlg))
print(confusion_matrix(y_test,predlg))
print(classification_report(y_test,predlg))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
y_pred_prob=lg.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1])
plt.plot(fpr,tpr,label='LogisticRegression')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('LogisticRegression')
plt.show()
```
